
Predictive Modeling with Linear Regression

1. Import resquired libraires

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

2. Load the dataset

```python
data = pd.read_csv('C:\\Users\\Rajkumar\\OneDrive\\Desktop\\New folder (2)\\JavaPrograms\\suhas datasets\\housing.csv')
data.head
```

```python
# Remove rows with missng values from the housing dataset
data_cleaned = data.dropna()
```

3. Data Preparation

```python
X = data_cleaned[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
Y = data_cleaned['median_house_value']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

4. Model Training

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

5. Model Evaluation

```python
Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

6. Plotting the Predicted Result

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting the actual vs predicted house prices
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred, color='blue', label='Actual vs Predicted')
plt.xlabel('Actual House Prices ($)')
plt.ylabel('Predicted House Prices ($)')
plt.title('Actual vs Predicted House Prices')

# Adding the regression line
x = np.linspace(min(Y_test), max(Y_test), 100)
y = x
plt.plot(x, y, color='red', label='Regression Line')
plt.legend()

plt.show()
```
